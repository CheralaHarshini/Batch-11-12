```python
# Step 1: Import libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create a sample dataset for 200 students
np.random.seed(42)  # for reproducibility

df = pd.DataFrame({
    'study_hours': np.random.normal(5, 1.5, 200).clip(0),  # mean=5 hours/day
    'sleep_hours': np.random.normal(7, 1, 200).clip(4, 10),  # mean=7 hours/day
    'gender': np.random.choice(['Male', 'Female'], 200)
})

# Let's assume exam_score depends on study_hours and sleep_hours
df['exam_score'] = (
    10 * df['study_hours']
    + 2 * df['sleep_hours']
    + np.random.normal(0, 10, 200)
).clip(0, 100)

# Step 3: Scatter plot with regression line
sns.lmplot(
    data=df,
    x='study_hours',
    y='exam_score',
    hue='gender',     # color by gender
    markers=['o', 's'],
    palette='coolwarm',
    height=6,
    aspect=1.2
)
plt.title("Relationship between Study Hours and Exam Score by Gender")
plt.show()

# Step 4: Pairplot for all relationships
sns.pairplot(
    data=df,
    vars=['study_hours', 'sleep_hours', 'exam_score'],
    hue='gender',
    kind='reg',
    diag_kind='kde'
)
plt.suptitle("Pairwise Relationships among Study, Sleep, and Exam Score", y=1.02)
plt.show()
```
