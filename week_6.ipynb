```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create sample blood sugar data for 200 patients
np.random.seed(42)
normal_patients = np.random.normal(100, 10, 120)   # normal range
diabetic_patients = np.random.normal(160, 15, 80)  # diabetic range
blood_sugar = np.concatenate([normal_patients, diabetic_patients])

df = pd.DataFrame({'blood_sugar': blood_sugar})

# KDE Plot
sns.kdeplot(data=df, x='blood_sugar', fill=True, color='teal', linewidth=2)
plt.title("Distribution of Blood Sugar Levels (KDE Plot)")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")
plt.show()
```
